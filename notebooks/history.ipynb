```python
import pandas as pd
```

```python
file_path = "../data/landing/Moving annual rent by suburb - March quarter 2023 copy.xlsx"
```

```python
df_1_bedroom_flat_cleaned = pd.read_excel(file_path, sheet_name='1 bedroom flat', header=[0, 1])
df_1_bedroom_flat_cleaned.columns = [f'{col1} {col2}'.strip() if pd.notna(col1) and pd.notna(col2) else col1 or col2 for col1, col2 in df_1_bedroom_flat_cleaned.columns]
df_1_bedroom_flat_cleaned = df_1_bedroom_flat_cleaned.drop([0])

df_1_bedroom_flat_cleaned.reset_index(drop=True, inplace=True)

df_1_bedroom_flat_cleaned.head()
```

```python
# Create separate DataFrames for Count and Median
count_df = df_1_bedroom_flat_cleaned[[col for col in df_1_bedroom_flat_cleaned.columns if '.1' not in col]+['Moving annual rent by suburb 1 bedroom flat.1']]
median_df = df_1_bedroom_flat_cleaned[[col for col in df_1_bedroom_flat_cleaned.columns if '.1' in col]+['Moving annual rent by suburb 1 bedroom flat.1']]

# Rename the columns in the Median DataFrame to remove the '.1'
median_df.columns = [col.replace('.1', '').strip() for col in median_df.columns]

# Optionally reset index to clean both DataFrames
count_df.reset_index(drop=True, inplace=True)
median_df.reset_index(drop=True, inplace=True)

count_df
```

```python
import re

def extract_date(col_name):
    match = re.findall(r'Lease commenced in year ending  (\w{3} \d{4})', col_name)
    if match:
        return match[0] 
    return col_name 

count_df = df_1_bedroom_flat_cleaned[[col for col in df_1_bedroom_flat_cleaned.columns if '.1' not in col] + ['Moving annual rent by suburb 1 bedroom flat.1']]
median_df = df_1_bedroom_flat_cleaned[[col for col in df_1_bedroom_flat_cleaned.columns if '.1' in col]+ ['Moving annual rent by suburb 1 bedroom flat.1']]
count_df.head(2)

```

```python
print(count_df.columns)
```

```python
count_df.columns = [extract_date(col) for col in count_df.columns]
count_df = count_df.rename({'Moving annual rent by suburb 1 bedroom flat': 'Suburb', 'Moving annual rent by suburb 1 bedroom flat.1':'Region'}, axis=1)
median_df.columns = [extract_date(col.replace('.1', '')) for col in median_df.columns]
median_df = median_df.rename({'Moving annual rent by suburb 1 bedroom flat': 'Suburb', 'Moving annual rent by suburb 1 bedroom flat.1':'Region'}, axis=1)

```
